```python
import pandas as pd
import skmultiflow  as sk
# https://scikit-multiflow.readthedocs.io/en/stable/installation.html

%matplotlib inline
```

```python
# https://scikit-multiflow.readthedocs.io/en/stable/api/generated/skmultiflow.data.AnomalySineGenerator.html#skmultiflow.data.AnomalySineGenerator

anomaly_stream = sk.data.AnomalySineGenerator(n_samples=10000, n_anomalies=250, contextual=True, n_contextual=50, noise = 0, replace=True, random_state=None)
```

```python
data, label = anomaly_stream.next_sample(10000)
```

```python
data_df = pd.DataFrame(data)
label_series = pd.Series(label)
```

```python
data_df[:200].plot(figsize=(20, 5))
```

```python
label_series.value_counts()
```

```python
dataset = pd.concat((data_df, label_series), axis=1)
dataset.to_csv('dummy_dataset.csv', index=False)
```
